# 클래스
- 초기화용 메서드 : 생성자(constructure)

```python
# 클래스
class Man:
    def __init__(self, name):
        self.name = name
        print("Initialized!")
        
    def hello(self):
        print("Hello " + self.name + "!")
    
    def goodbye(self):
        print("Good-bye " + self.name + "!")
    

m = Man("David")
m.hello()
m.goodbye()
        
```

# Nunpy
- 넘파이 산술연산 : 두 배열의 원소 수가 같아야 한다
- 브로드캐스트 : 형상이 다른 배열끼리의 계산 (스칼라 값이 크기에 맞게 확대된 후 계산됨)

```python
import numpy as np

x = np.array([1.0, 2.0, 3.0])
y = np.array([2.0, 4.0, 6.0])
x + y
```

```python
# 브로트캐스트
A = np.array([[1,2],[3,4]])
B = np.array([10,20])
A*B
```

```python
# 원소 접근
X = np.array([[51,55],[14,19],[0,4]])
print(X)
X[0]
```

```python
# for문을 이용한 원소 접근
for row in X:
    print(row)
```

```python
# 인덱스를 사용하여 배열 접근
X = X.flatten() # X를 1차원 배열로 변환(평탄화)
print(X)
X[np.array([0,2,4])]
```

```python
# 특정 조건을 만족하는 원소만 접근
print(X>15)
print(X[X>15])
```

# matplotlib
- 그래프를 그리기 위한 라이브러리(데이터 시각화)

```python
# 단순한 그래프 그리기
import numpy as np
import matplotlib.pyplot as plt

# 데이터 준비
x = np.arange(0, 6, 0.1)     # 0에서 6까지 0.1 간격으로 생성
y = np.sin(x)

# 그래프 그리기
plt.plot(x,y)
plt.show()
```

```python
# pyplot의 기능
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='sin')
plt.plot(x, y2, linestyle="--", label='cos')  # cos 함수 점선으로 그리기
plt.xlabel("x")  # x축 이름
plt.ylabel("y")  # y축 이름
plt.title('sin & cos')  # 제목
plt.legend(loc=(0.8, 0.8))  # loc : label 위치 조정(0,0 : 왼쪽 아래 1,1 : 오른쪽 위)
plt.show()
```
